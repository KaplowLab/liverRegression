```python
import pandas as pd
import numpy as np

filename='macaqueLCL_pos_ALL.bed'
dr = '/home/azstephe/immunetacit/repos/lcl_cnn/data'
log_same_size = pd.read_csv(f'{dr}/random_log_10525/macaqueLCL_pos_ALL.bed', header=None, sep='\t')
log_all = pd.read_csv(f'{dr}/log/macaqueLCL_pos_ALL.bed', header=None, sep='\t')
log_sorted = pd.read_csv(f'{dr}/log_sorted/macaqueLCL_pos_ALL.bed', header=None, sep='\t')
qn = pd.read_csv(f'{dr}/random_log_10525/quantile_norm/macaqueLCL_pos_ALL.bed', header=None, sep='\t')
notIncluded = macAll[20615:]

sortedIncluded = log_sorted[~log_sorted[3].isin(notIncluded[3])].reset_index(drop=True)

notIncluded.columns = range(notIncluded.shape[1])
sortedIncluded.columns = range(log_sorted.shape[1])

def find_closest(row, sortedIncluded):
    closest_idx = (sortedIncluded[4] - row[4]).abs().idxmin() # get index with closest value
    return sortedIncluded.loc[closest_idx, 3] # keep peak name of that index

# adds column with closest peak in the included random list
notIncluded = notIncluded.copy()
notIncluded.loc[:, 'closest_col3'] = notIncluded.apply(lambda row: find_closest(row, sortedIncluded), axis=1)
# not Included has the closest peakName

# clean up the df 
df1 = notIncluded.copy()
df2 = qn.copy()
df1 = df1.merge(df2[[3, 4]], left_on='closest_col3', right_on=3, how='left')

df1.drop(columns=['closest_col3', '3_y', '4_x'], inplace=True)
df1.rename(columns={'3_x': 3, '4_y': 4}, inplace=True)

# merge to the final concatenated df with all peaks and their associated signal values
ladderQN = pd.concat([macQN, df1]).reset_index(drop=True)
# ladderQN.to_csv(f'{dr}/ladder_qn/{filename}', header=None, index=False, sep='\t')
ladderQN
```
