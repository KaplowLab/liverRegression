```python
# getting true mouse weak, medium, strong peaks
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

species='mouse'

val_df = pd.read_csv(f'/home/azstephe/liverRegression/regression_liver/data/splits/logPos/{species}_liver_VAL.narrowPeak', header=None, delim_whitespace=True)
train_df = pd.read_csv(f'/home/azstephe/liverRegression/regression_liver/data/splits/logPos/{species}_liver_TRAINONLY.narrowPeak', header=None, delim_whitespace=True)

true_pos_df = pd.concat([val_df, train_df]).reset_index(drop=True)
true_pos_df
```

```python
# 2p5 read as 2 point 5
greater_than_2p5 = true_pos_df[true_pos_df[4]>2.5][3].to_frame()
greater_than_2p5.to_csv('/home/azstephe/liverRegression/regression_liver/data/mouse_strength/mouse_greater_than_2p5_peaklist.csv', sep='\t', header=None, index=False)

less_than_1p4 = true_pos_df[true_pos_df[4]<1.4][3].to_frame()
less_than_1p4.to_csv('/home/azstephe/liverRegression/regression_liver/data/mouse_strength/mouse_less_than_1p4_peaklist.csv', sep='\t', header=None, index=False)

between_1p4_2 = true_pos_df[(true_pos_df[4]>1.4) & (true_pos_df[4]<2)][3].to_frame()
between_1p4_2.to_csv('/home/azstephe/liverRegression/regression_liver/data/mouse_strength/mouse_between_1p4_2_peaklist.csv', sep='\t', header=None, index=False)
```
