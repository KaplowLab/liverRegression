```python
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
def histogram(pred_df, name, status):
    plt.hist(pred_df, bins=50)
    plt.xlim(0,4.1)
    plt.xlabel(f'{name}')
    plt.ylabel('Count')
    plt.title(f'Histogram of {status}')
    plt.show()
```
