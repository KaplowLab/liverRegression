```python
import numpy as np
import matplotlib.pyplot as plt

points = [(10, 10), (20, 20), (30, 30)]
theta = np.linspace(0, np.pi, 180)

plt.figure(figsize=(8, 6))
for x, y in points:
    rho = x * np.cos(theta) + y * np.sin(theta)
    plt.plot(theta, rho, label=f"({x},{y})")

plt.xlabel(r"$\theta$")
plt.ylabel(r"$\rho$")
plt.title("Hough Transform Sinusoids")
plt.legend()
plt.grid(True)

plt.show()

x1, y1 = points[0]
x2, y2 = points[1]

m = (y2 - y1) / (x2 - x1)
c = y1 - m * x1

# Print the result
print(f"m = {m:.2f}, c = {c:.2f}")
```
